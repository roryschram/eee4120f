## White Noise Generation

```python
plt.plot(df1['Sample'], df1['Execution time of Method 1 (ms)']) 
plt.scatter(df1['Sample'], df1['Execution time of Method 1 (ms)']) 
plt.tight_layout()
plt.xlabel("Sample Size")
plt.ylabel("Execution Time (ms)")
plt.show()
```

```python
plt.plot(df1['Sample'], df1['Execution time of Method 2 (ms)']) 
plt.scatter(df1['Sample'], df1['Execution time of Method 2 (ms)']) 
plt.tight_layout()
plt.xlabel("Sample Size")
plt.ylabel("Execution Time (ms)")
plt.show()
```
